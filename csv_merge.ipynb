```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
rows = {
  "L_ICA": [],
  "R_ICA": [],
  "L_PCOM": [],
  "R_PCOM": [],
  "L_AntChor": [],
  "R_AntChor": [],
  "L_ACA": [],
  "R_ACA": [],
  "L_ACOM": [],
  "R_ACOM": [],
  "L_MCA": [],
  "R_MCA": [],
  "L_VA": [],
  "R_VA": [],
  "L_PICA": [],
  "R_PICA": [],
  "L_SCA": [],
  "R_SCA": [],
  "BA": [],
  "L_PCA": [],
  "R_PCA": [],
}
```

```python
row_dict = defaultdict(lambda: [0] * 1127)
```

```python
name_lst = [
  "testLI-A.csv",
  "testLI-B.csv",
  "testLV-A.csv",
  "testLV-B.csv",
  "testRI-A.csv",
  "testRI-B.csv",
  "testRV-A.csv",
  "testRV-B.csv",
]
for name in name_lst:
  df = pd.read_csv(name)
  for i, row in df.iterrows():
    for k, v in row.items():
      v = round(v, 4)
      row_dict[k][i] = max(row_dict[k][i], v)
```

```python
result = pd.DataFrame.from_dict(data=row_dict, orient='columns')
```
